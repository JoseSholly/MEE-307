```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style= "darkgrid")
```

# Vertical Pipe

```python
data= pd.read_csv('Mee_V_data.csv', index_col= ["Fluid", "Pipe Length"])
```

```python
data.head(60)
```

```python
gfg=sns.relplot(x= 'Reynold', y='Head Loss',kind= 'line',hue='Fluid', data= data, palette= 'viridis')

gfg.set(xlabel='Reynold\'s Number (R\u0065)', ylabel= 'Head Loss', title='Vertical Pipe(Gold)')
```

```python
gfg=sns.relplot(x= 'Pipe Length', y='Reynold',kind= 'line',hue='Fluid', data= data, palette= 'viridis')

gfg.set(xlabel='Pipe Length', ylabel= "Reynold\'s Number (R\u0065)", title='Vertical Pipe(Gold)')
```

```python
gfg=sns.relplot(x= 'Reynold', y='Coefficient',kind= 'line',hue='Fluid', data= data, palette= 'viridis')

gfg.set(xlabel='Reynold\'s Number (R\u0065)', 
            ylabel= 'Co-efficient of Heat Transfer',  title='Vertical Pipe(Gold)')
```

```python
gfg=sns.relplot(x= 'Reynold', y='Pressure',kind= 'line',hue='Fluid', data= data, palette= 'viridis')

gfg.set(xlabel='Reynold\'s Number (R\u0065)', 
            ylabel= 'Pressure', title='Vertical Pipe(Gold)')
```
